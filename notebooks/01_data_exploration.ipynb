```
# Data Exploration Notebook
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.data_loader import load_biometric_data, load_triage_data

# Load data
biometric_df = load_biometric_data('../data/sample_biometric.csv')
triage_df = load_triage_data('../data/sample_triage.csv')

# Display basic info
print("Biometric Data:")
print(biometric_df.info())
print("\nTriage Data:")
print(triage_df.info())

# Show basic statistics
print("\nBiometric Data Statistics:")
print(biometric_df.describe())
print("\nTriage Data Statistics:")
print(triage_df.describe())

# Visualize data distribution
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
biometric_df['citizenship_status'].value_counts().plot(kind='bar')
plt.title('Citizenship Status Distribution')

plt.subplot(1, 2, 2)
triage_df['triage_priority'].value_counts().plot(kind='bar')
plt.title('Triage Priority Distribution')

plt.tight_layout()
plt.show()
```
